```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#note to reviewer, i just wanted to also see basic stats in addition to the request in the bullet to help with insights, 
#and i often have extra code in here bc i was checking things out

#stats = purchase_data.describe()
#stats

unique_items = len(purchase_data["Item ID"].unique())
unique_items

average_price = purchase_data["Price"].mean()
average_price

number_of_purchases = purchase_data["Item ID"].count()
number_of_purchases

total_revenue = purchase_data["Price"].sum()
total_revenue

purchase_analysis = pd.DataFrame ({"Number of Unique Items" : [unique_items],
                                   "Average Price" : average_price,
                                   "Number of Purchases" : number_of_purchases,
                                   "Total Revenue" : total_revenue})


# Use Map to format all the columns
purchase_analysis["Average Price"] = purchase_analysis["Average Price"].map("${:.2f}".format)
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].map("${:,.0f}".format)

purchase_analysis
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Slice the data and place it into bins
pd.cut(purchase_data["Age"], bins, labels=group_names).head()

# Place the data series into a new column inside of the DataFrame
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names)

#groupbyhere 

age_demo_groupby = purchase_data.groupby(["Age Group"])
#age_demo_groupby

unique_items = age_demo_groupby["Item ID"].count()
#unique_items

total_items = age_demo_groupby["Item ID"].sum()

avg_purchase_price = age_demo_groupby["Price"].mean()
#avg_purchase_price

total_value = age_demo_groupby["Price"].sum()
#total_value

avg_total_purchase = (100*total_value) / total_items
#avg_total_purchase

purchase_analysis_age = pd.DataFrame ({"Purchase Count" : unique_items,
                                         "Average Purchase Price" : avg_purchase_price, 
                                         "Total Purchase Value" : total_value,
                                         "Avg Total Purchase per Person" : avg_total_purchase})


# Use Map to format all the columns
purchase_analysis_age["Average Purchase Price"] = purchase_analysis_age["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis_age["Total Purchase Value"] = purchase_analysis_age["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_age["Avg Total Purchase per Person"] = purchase_analysis_age["Avg Total Purchase per Person"].map("${:.2f}".format)
purchase_analysis_age

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders = purchase_data.groupby(["SN"])

unique_items = top_spenders["Item ID"].count()
#unique_items

avg_purchase_price = top_spenders["Price"].mean()
#avg_purchase_price

total_value = top_spenders["Price"].sum()
#total_value

top_spenders = pd.DataFrame ({"Purchase Count" : unique_items,
                                "Average Purchase Price" : avg_purchase_price, 
                                "Total Purchase Value" : total_value
                                 })

list_of_top = top_spenders.sort_values("Purchase Count", ascending = False)

list_of_top["Average Purchase Price"] = list_of_top["Average Purchase Price"].map("${:.2f}".format)
list_of_top["Total Purchase Value"] = list_of_top["Total Purchase Value"].map("${:.0f}".format)
list_of_top.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#columns =["Item ID", "Item Name", "Item Price"]

most_pop = purchase_data.groupby(['Item ID', 'Item Name'])

purchase_count = most_pop["Item ID"].count()
#purchase_count

total_purchase_value = most_pop["Price"].sum()
#total_purchase_value

item_price = total_purchase_value / purchase_count
#item_price

most_pop = pd.DataFrame ({"Purchase Count" : purchase_count,
                          "Item Price" : item_price,
                          "Total Purchase Value" : total_purchase_value
                         })

most_pop_sort = most_pop.sort_values("Purchase Count", ascending=False)

most_pop_sort["Item Price"] = most_pop["Item Price"].map("${:.2f}".format)
most_pop_sort["Total Purchase Value"] = most_pop["Total Purchase Value"].map("${:.0f}".format)
most_pop_sort
```
